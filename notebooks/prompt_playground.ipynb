```python
# ---------------------------------------------------
# Prompt Playground â€” Jupyter Lab
# Author: Zayed
# ---------------------------------------------------

from openai import OpenAI
import pandas as pd
import os

client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

# -----------------------------
# 1. Utility functions
# -----------------------------
def run_prompt(prompt, model="gpt-4o-mini", temperature=0.2, max_tokens=300):
    response = client.chat.completions.create(
        model=model,
        messages=[{"role": "user", "content": prompt}],
        temperature=temperature,
        max_tokens=max_tokens
    )
    return response.choices[0].message.content


# -----------------------------
# 2. Prompt Styles
# -----------------------------
def zero_shot(task):
    return task

def few_shot(task, examples):
    return examples + "\n\n" + "Task:\n" + task

def chain_of_thought(task):
    return f"{task}\n\nThink step by step and show reasoning."

def socratic_style(task):
    return f"You are a Socratic tutor. Ask guided questions.\n\n{task}"


# -----------------------------
# 3. Run comparisons
# -----------------------------
# Load few-shot examples
with open("../prompts/examples/reasoning_fewshot.txt") as f:
    reasoning_examples = f.read()

with open("../prompts/examples/classification_fewshot.txt") as f:
    classification_examples = f.read()

with open("../prompts/examples/coding_fewshot.txt") as f:
    coding_examples = f.read()


task = "If a tree is taller than a house, and the house is taller than a car, what is the tallest?"

prompts = {
    "zero_shot": zero_shot(task),
    "few_shot_reasoning": few_shot(task, reasoning_examples),
    "few_shot_classification": few_shot(task, classification_examples),
    "few_shot_coding": few_shot(task, coding_examples),
    "chain_of_thought": chain_of_thought(task),
    "socratic": socratic_style(task),
}

results = []

for mthd_name, p in prompts.items():
    output = run_prompt(p)
    results.append({
        "method": mthd_name,
        "Prompt": p,
        "output": output
    })

df = pd.DataFrame(results)

df


```
